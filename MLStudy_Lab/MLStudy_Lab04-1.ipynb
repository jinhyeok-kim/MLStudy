## Hypothesis using matrix

variable이 많을 때 Y를 예측하는 모델의 경우

$$H(x_1, x_2, x_3) = x_1w_1 + x_2w_2 + x_3w_3 $$

```python
import tensorflow as tf
import numpy as np

xy = np.loadtxt('data-01-test-score.csv', delimiter=',', dtype=np.float32)

x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]

x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]

# placeholdlers
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random_normal([1]), name='weight1')
w2 = tf.Variable(tf.random_normal([1]), name='weight2')
w3 = tf.Variable(tf.random_normal([1]), name='weight3')
b = tf.Variable(tf.random_normal([1]), name='bias')

hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b

# cost/Los function
# 이 전에 쓰던 것과 동일함
cost = tf.reduce_mean(tf.square(hypothesis - Y))

#Minimize learning_rate를 낮게 줌
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)


sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    #train을 돌리고 그 때, cost와 hypothes를 같이 기록을 해서 출력을 하는 모델
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict={x1: x1_data, x2: x2_data, x3: x3_data, Y: y_data})
    if step % 10 == 0:
        print step ,'Cost : ', cost_val,"\n Prediction: \n", hy_val,"\n"
              
```

<br>
위의 코드의 경우 variable이 너무 많아 복잡해져 스파게티 코드가 될 가능성이 **매우 높다!!**<br>

## Hypothesis using matrix

<br>

$$\quad
\begin{pmatrix} 
x_1 & x_2 & x_3 
\end{pmatrix}
\quad
\cdot
\quad
\begin{pmatrix} 
w_1 \\
w_2 \\
w_3 \\
\end{pmatrix}
\quad
=
\quad
\begin{pmatrix} 
x_1w_1 + x_2w_2 + x_3w_3
\end{pmatrix}
\quad$$
<br><br>
$$ H(X) = XW $$

```python
import tensorflow as tf

# X_data []하나가 x1, x2 ... 각각은 instance
x_data = [[73., 80., 75.], [93., 88., 93.], [89., 91., 90.], [96., 91., 90.], [73., 66., 70.]]
y_data = [[152.], [185.], [180.], [196.], [142.]]

# placeholdlers
# shape을 주의해줘야한다. 위의 경우 x1, x2, x3 세개이므로 shape 뒤의 값은 3개다.
# shape 앞에 값은 Data의 갯순데 갯수는 추가하기 나름이므로 정해져 있지 않는 경우 None으로 표현한다.
# [x1, x2, x3]가 n개 있을 수 있으므로 [None, 3]

X = tf.placeholder(tf.float32, shape=[None, 3])
Y = tf.placeholder(tf.float32, shape=[None, 1])

W = tf.Variable(tf.random_normal([3, 1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

# hypothesis
hypothesis = tf.matmul(X, W) + b

# cost/Los function
# 이 전에 쓰던 것과 동일함
cost = tf.reduce_mean(tf.square(hypothesis - Y))

#Minimize learning_rate를 낮게 줌
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(cost)


sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    #train을 돌리고 그 때, cost와 hypothes를 같이 기록을 해서 출력을 하는 모델
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict={X: x_data, Y: y_data})
    if step % 10 == 0:
        print step ,'Cost : ', cost_val,"\n Prediction: \n", hy_val,"\n"
              
```
